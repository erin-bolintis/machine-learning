```python
import os
from PIL import Image
import pandas as pd

# Get the current working directory (where your notebook is located)
current_directory = os.getcwd()

# Construct the path to the 'data' folder
data_folder = os.path.join(current_directory, 'data')

# Construct the path to the 'posters' folder
posters_folder = os.path.join(data_folder, 'posters')

#all the path of the movie folders
image_directory_posters_2000 = os.path.join(posters_folder, '2000')
image_directory_posters_2001 = os.path.join(posters_folder, '2001')
image_directory_posters_2002 = os.path.join(posters_folder, '2002')

def get_image_info(image_path):
    with Image.open(image_path) as img:
        width, height = img.size
        size = os.path.getsize(image_path)
        pixels = width * height
    return {'filename': os.path.basename(image_path),
            'size': size,
            'width': width,
            'height': height,
            'pixels': pixels}

# Specify the directory for the year 2000
image_directory = image_directory_posters_2000

image_info_list = []

for filename in os.listdir(image_directory):
    if filename.endswith('.jpg') or filename.endswith('.png'):
        image_path = os.path.join(image_directory, filename)
        image_info = get_image_info(image_path)
        image_info_list.append(image_info)

# Set display options to show all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

image_df = pd.DataFrame(image_info_list)
image_df['movie_year'] = 2000
df_2000 = image_df.sort_values(by=['width', 'height'], ascending=[True, True])
df_2000

```
